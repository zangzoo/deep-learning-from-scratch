### 1.4.2 클래스
- 직접 클래스를 정의하면 독자적인 자료형을 만들 수 있음
- 클래스에는 그 클래스만의 전용 함수(메서드)와 속성 정의 가능

```python
class Man:
  # 클래스의 인스턴스가 만들어질 때 한 번만 불리며, 메서드의 첫 번째 인수로 자신(self)를 명시적으로 쓰는 것이 특징
  def __init__(self, name): # 생성자: 클래스를 초기화하는 방법 정의
    # 인수인 name으로 인스턴스 변수인 self.name을 초기화
    self.name = name
    print("Initialized!")

  def hello(self):  # 메서드 1
    print("Hello " + self.name + "!")

  def goodbye(self):  # 메서드 2
    print("Good-bye " + self.name + "!")

m = Man("David")
m.hello()
m.goodbye()
```

## 1.6 matplotlib
- 그래프를 그려주는 라이브러리 + 데이터 시각화

### 1.6.1 단순한 그래프 그리기
- 그래프를 그리려면 matplotlib의 pyplot 모듈을 이용함

```python
# sin 함수 그리기

import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성 [0, 0.1, 0.2, ..., 5.8, 5.9]
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
```

### 1.6.2 pyplot의 기능

```python
# cos 함수

import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # cos 함수는 점선으로 그리기
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title('sin & cos') # 제목
plt.legend()  # 범례 표시
plt.show()
```
